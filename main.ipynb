```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
```

```python
audio_file = 'crowd_train/wavs/475e76f77ac1ed7cabafca740b15b32a.wav'
```

```python
signal, sample_rate = librosa.load(audio_file)

signal.shape
```

```python
mfccs = librosa.feature.mfcc(y=signal, n_mfcc=13, sr=sample_rate)
mfccs.shape
```

```python
plt.figure(figsize=(6, 3))
librosa.display.specshow(mfccs, x_axis="time", sr=sample_rate)

plt.colorbar(format='%+2f')
plt.show()
```
