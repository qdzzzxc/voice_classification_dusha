```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.random import normal
from numpy.random import uniform
from soinn import Soinn
from esoinn import ESoinn
from random import choice

from sklearn.cluster import KMeans
from sklearn.metrics import adjusted_rand_score

# generate data
n = 55_000
sigma = 0.02
c = 10.0 * np.random.rand(n) - 5.0
x0 = -7.05
y0 = 0.0
x1 = -7.30
y1 = 0.0
Y = [[normal(x0, sigma), normal(y0, sigma)] for i in range(int(len(c)/4))]
X = [[c[i]/10 - 6.25, np.sin(c[i])/20+uniform(-0.015, 0.015)] for i in range(len(c))]
Z = [[normal(x1, sigma), normal(y1, sigma)] for i in range(int(len(c)/4))]
Y_l = [0] * int(len(c)/4)
X_l = [1] * int(len(c))
Z_l = [2] * int(len(c)/4)
Y.extend(X)
Y.extend(Z)
X = np.array(Y)
Y_l.extend(X_l)
Y_l.extend(Z_l)
X_l = np.array(Y_l)
```

```python
len(X_l), len(X)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=X_l)
```

```python
sk_kmeans = KMeans(n_clusters=3, n_init='auto', random_state=0)
sk_kmeans.fit(X)
```

```python
sk_kmeans_pred_res = sk_kmeans.predict(X)
sk_kmeans_ari = adjusted_rand_score(X_l, sk_kmeans_pred_res)
sk_kmeans_centroinds = sk_kmeans.cluster_centers_
print(f'Adjusted Rand Score for sk KMeans: {sk_kmeans_ari}', '', sep='\n')
print(sk_kmeans_centroinds, '', sep='\n')
print('prediction', sk_kmeans_pred_res, sep='\n')
```

```python
plt.scatter(X[:, 0], X[:, 1], c=X_l)
plt.scatter(sk_kmeans_centroinds[:, 0], sk_kmeans_centroinds[:, 1], c='red')
```
